```python
WINDOW_SIZE = 12
nsei['rolling_mean'] = nsei.Close.rolling(window=WINDOW_SIZE).mean()
nsei['rolling_std'] = nsei.Close.rolling(window=WINDOW_SIZE).std()
```

```python
corr_close_mstd = nsei.rolling_mean.corr(nsei.rolling_std)
corr_close_mstd
```

```python
nsei[['rolling_mean','rolling_std']].plot(subplots=True,figsize=(12,6))
```

```python
WINDOW_SIZE = 12
nsei['Vrolling_mean'] = nsei.Volume.rolling(window=WINDOW_SIZE).mean()
nsei['Vrolling_std'] = nsei.Volume.rolling(window=WINDOW_SIZE).std()
```

```python
nsei[['Vrolling_mean','Vrolling_std']].plot(subplots=True,figsize=(12,6))
```

```python
nsei[['Vrolling_mean','rolling_std']].plot(subplots=True,figsize=(12,6))
```

```python
corr_vmean_mstd = nsei.Vrolling_mean.corr(nsei.rolling_std)
corr_vmean_mstd
```

```python
nsei[['Vrolling_std','rolling_std']].plot(subplots=True,figsize=(12,6))
```

```python
corr_cstd_mstd = nsei.Vrolling_std.corr(nsei.rolling_std)
corr_cstd_mstd
```

```python
nsei['logrolling_mean']=np.log(nsei.rolling_mean/nsei.rolling_mean.shift(1))
nsei['logrolling_std']=np.log(nsei.rolling_std/nsei.rolling_std.shift(1))
```

```python
nsei['logVrolling_mean']=np.log(nsei.Vrolling_mean/nsei.Vrolling_mean.shift(1))
nsei['logVrolling_std']=np.log(nsei.Vrolling_std/nsei.Vrolling_std.shift(1))
```

```python
import numpy as np
import yfinance as yf
import seaborn as sns
import scipy.stats as scs
import statsmodels.api as sm
import statsmodels.tsa.api as smt
```

```python
r_range = np.linspace(min(nsei.Close_log_rtn), max(nsei.Close_log_rtn), num=1000)
mu = nsei.Close_log_rtn.mean()
sigma = nsei.Close_log_rtn.std()
norm_pdf = scs.norm.pdf(r_range, loc=mu, scale=sigma)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
# histogram
sns.distplot(nsei.Close_log_rtn, kde=False, norm_hist=True, ax=ax[0])
ax[0].set_title('Distribution of NSEI returns', fontsize=16)
ax[0].plot(r_range, norm_pdf, 'g', lw=2,
label=f'N({mu:.2f}, {sigma**2:.4f})')
ax[0].legend(loc='upper left');
# Q-Q plot
qq = sm.qqplot(nsei.Close_log_rtn.values, line='s', ax=ax[1])
ax[1].set_title('Q-Q plot', fontsize = 16)
```

```python
N_LAGS = 50
SIGNIFICANCE_LEVEL = 0.05
```

```python
acf = smt.graphics.plot_acf(nsei.Close_log_rtn,
lags=N_LAGS,
alpha=SIGNIFICANCE_LEVEL
)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 10))
smt.graphics.plot_acf(nsei.Close_log_rtn ** 2, lags=N_LAGS,
alpha=SIGNIFICANCE_LEVEL, ax = ax[0])
ax[0].set(title='Autocorrelation Plots',
ylabel='Squared Returns')
smt.graphics.plot_acf(np.abs(nsei.Close_log_rtn), lags=N_LAGS,
alpha=SIGNIFICANCE_LEVEL, ax = ax[1])
ax[1].set(ylabel='Absolute Returns',
xlabel='Lag')
```

```python
nsei['moving_std_252'] = nsei[['Close_log_rtn']].rolling(window=252).std()
nsei['moving_std_21'] = nsei[['Close_log_rtn']].rolling(window=21).std()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(18, 15),
sharex=True)
nsei.plot(ax=ax[0])
ax[0].set(title='NSEI time series',
ylabel='Stock price ($)')
nsei.Close_log_rtn.plot(ax=ax[1])
ax[1].set(ylabel='Close_log_rtn (%)',xlabel='Date')
ax[1].legend()
nsei.moving_std_252.plot(ax=ax[2], color='r',
label='Moving Volatility 252d')
nsei.moving_std_21.plot(ax=ax[2], color='g',
label='Moving Volatility 21d')
ax[2].set(ylabel='Moving Volatility',
xlabel='Date')
ax[2].legend()
```
