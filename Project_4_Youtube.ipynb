```python
data=data.replace('--',np.nan,regex=True) #Replace -- into nan
data.head(20)
```



Check null value in the Dataset



```python
data.isnull()
```

```python
per_missing = data.isnull().sum()*100/ len(data)  #get missing value into percent
```

```python
data.dropna(axis=0,inplace=True)
sns.heatmap(data.isnull())
```

Data cleaning [Rank column]

```python
data.head()
```

```python
data.tail(5)
```

```python
data.dtypes
```

```python
data['Rank']=data['Rank'].str[0:-2]
data.head(5)
```

```python
data['Rank']=data['Rank'].str.replace(',','')
```

```python
data.tail()
```

```python
data['Rank']=data['Rank'].str.replace(',','').astype('int')
```

```python
data['Subscribers']=data['Subscribers'].astype('int')
```
