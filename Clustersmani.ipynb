```python
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import AffinityPropagation
from matplotlib import pyplot
```

```python
# define dataset
X, _ = make_classification(n_samples=1300, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
# define the model
model = AffinityPropagation(damping=0.9)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
	# get row indexes for samples with this cluster
	row_x = where(yhat == cluster)
	# create scatter of these samples
	pyplot.scatter(X[row_x, 0], X[row_x, 1])
# show the plot
pyplot.show()
```

Scatter Plot of Dataset With Clusters Identified Using Affinity Propagation.
Real-valued messages are exchanged between data points until a high-quality set of exemplars and corresponding clusters gradually emerges.

**Agglomerative Clustering**
means desired number of clustrs achived

```python

# agglomerative clustering
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import AgglomerativeClustering
from matplotlib import pyplot
# define dataset
X, _ = make_classification(n_samples=1300, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
# define the model
model = AgglomerativeClustering(n_clusters=2)
# fit model and predict clusters
yhat = model.fit_predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
	# get row indexes for samples with this cluster
	row_x = where(yhat == cluster)
	# create scatter of these samples
	pyplot.scatter(X[row_x, 0], X[row_x, 1])
# show the plot
pyplot.show()
```

** *gaussian mixture clustering*

```python

from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.mixture import GaussianMixture
from matplotlib import pyplot
# define dataset
X, _ = make_classification(n_samples=1300, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
# define the model
model = GaussianMixture(n_components=2)
# fit the model
model.fit(X)
# assign a cluster to each example
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
	# get row indexes for samples with this cluster
	row_x = where(yhat == cluster)
	# create scatter of these samples
	pyplot.scatter(X[row_x, 0], X[row_x, 1])
# show the plot
pyplot.show()
```
