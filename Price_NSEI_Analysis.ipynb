```
import pandas as pd
import numpy as np
import yfinance as yf
```

```
nsei = yf.download('^NSEI',
 start='2020-01-01',
 end='2021-07-10',
 progress=False)
```

```
nsei['simple_rtn'] = nsei.Close.pct_change()
```

```
nsei['log_rtn'] = np.log(nsei.Close/nsei.Close.shift(1))
```

```
nsei['log_rtn'].plot(subplots=True, style='b',
figsize=(12, 8))
```
